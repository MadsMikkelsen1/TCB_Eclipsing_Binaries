## Script a PDM and perform a period determination, using trial periods between 30 and 80 days. Argue your choice of binning of the trial periods in this range. 

Related to num_bins:

If a large number of bins are used, we can accurately measure slight variations in the curves, which is useful if the curves show marked variations. The disadvantage is that if the data are widely scattered, the result is less robust. On the other hand, if a low number is chosen, we eliminate this possibility by grouping more values per bin, but we lose efficiency when finding periods in which the star varies rapidly.

```python

def pdm(times, fluxes, test_periods, num_bins=50):

    # Precompute the overall variance of the flux
    flux_variance = np.var(fluxes)

    # Initialize theta array
    theta_statistics = []

    # Iterate over each trial period
    for trial_period in test_periods:
        # Compute phases by folding on the trial period
        phases = np.mod(times, trial_period) / trial_period
        
        # Combine phases and fluxes for sorting and binning
        sorted_indices = np.argsort(phases)
        sorted_phases = phases[sorted_indices]
        sorted_fluxes = fluxes[sorted_indices]
        
        # Define bin edges and assign each phase to a bin
        bin_edges = np.linspace(0, 1, num_bins + 1)
        bins = np.digitize(sorted_phases, bin_edges) - 1
        
        # Calculate variance within each bin
        bin_variances = []
        for bin_num in range(num_bins):
            bin_mask = bins == bin_num
            bin_data = sorted_fluxes[bin_mask]
            
            if len(bin_data) > 1:  # Ensure the bin has sufficient data points
                bin_variances.append(np.var(bin_data))
        
        # Calculate the theta statistic for this period
        if bin_variances:
            theta = np.mean(bin_variances) / flux_variance
        else:
            theta = 1.0  # Default value if no valid bins
        
        theta_statistics.append(theta)

    return np.array(theta_statistics)
```

```python
# Define a range of trial periods
min_period = 30 # Minimum trial period (days)
max_period = 80 # Maximum trial period (days)
num_periods = 100 # Number of trial periods to test
trial_periods = np.linspace(min_period, max_period, num_periods)
# Perform PDM analysis
theta_values =  pdm(time_values, flux_values, trial_periods,num_bins=50)
# Plot the PDM theta values as a function of the trial period
plt.figure(figsize=(10, 10))
plt.plot(trial_periods, theta_values, label='PDM Theta',color='blue')
plt.title('PDM Analysis')
plt.xlabel('Trial Period [days]')
plt.ylabel('Theta Statistic')
```
